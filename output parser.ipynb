```python
import re
import pandas as pd
```

```python
filename = "log.txt"
line_number = 1
data = list()
travis = True

with open(filename) as f:
    for line in f:
        line = line.rstrip("\n")
        if "_ test_" in line:
            if travis:
                line = line[9:-8]
            try:
                test, case = line.strip("_").strip().split("[")
                case = case.rstrip("]")
            except:
                test = line.strip("_")
                case = ""
            data.append([test, case, "Line {}".format(str(line_number))])

        line_number += 1
```

```python
df = pd.DataFrame(data, columns=["test", "case", 'Line'])#
df.to_csv('failed.csv')
```
