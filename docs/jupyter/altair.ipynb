# Demonstrating Altair Modin Interoperability
### Each visualization is created with a Modin and then pandas dataframe for comparison.

```python
import modin.pandas as pd
import altair as alt
```

```python
from vega_datasets import data
cars_pandas = data.cars()
cars = pd.DataFrame(data.cars())
```

```python
# Create a visualization with Modin df
alt.Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).interactive()
```

```python
# Create a visualization with pandas df
alt.Chart(cars_pandas).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).interactive()
```
